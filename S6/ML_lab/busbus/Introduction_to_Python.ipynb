```python
largest = None
smallest = None
print("Before")
while True:
  num = input("Enter a number: ")
  if num == "done" :
    break
  try:
    print(num)
  except:
    print("Invalid number")
    continue

  num = input("Enter a number:")
  if smallest is None :
    smallest = num
  elif num < smallest :
    smallest = num
  if largest is None :
	  largest = num
  elif num > largest :
    largest = num
print("Maximum", largest)
print("Minimum", smallest)
```

#### Comparison between Vectorised and  Non Vectorised Implementation in Python

```python
import numpy as np
num = 50000000
a = np.random.random(num)
b = np.random.random(num)
type(a)
```

#### Vectorised Implementation

```python
import time
start = time.time()
c = np.dot(a,b)
end = time.time()
print(end)
print(c)
print("Time taken for Verctorized Implementation  : " +  str((end -start)*1000) + 'ms')
```

#### Non Vectorised Implementation (Looping)

```python
c =0
start = time.time()
for i in range(num):
    c += a[i]*b[i]
end = time.time()
print(c)
print("Time taken for Loop version  : " +  str((end -start)*1000) + 'ms')
```

#### Changing the Image Color Format

*  OpenCV reads in images in BGR format (instead of RGB) because when OpenCV was first being developed, BGR color format was popular among camera manufacturers and image software providers. The red channel was considered one of the least important color channels, so was listed last, and many bitmaps use BGR format for image storage. However, now the standard has changed and most image software and cameras use RGB format, which is why, in these examples, it's good practice to initially convert BGR images to RGB before analyzing or manipulating them.








#### Splitting the Image Color Format

```python
plt.figure(figsize=(10,10))

copy  = np.copy(img)
copy1 = cv2.cvtColor(copy,cv2.COLOR_BGR2RGB)

plt.subplot(121).imshow(copy)
plt.subplot(122).imshow(copy1)
```
